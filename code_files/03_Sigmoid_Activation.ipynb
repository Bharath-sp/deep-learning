```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(w,b,x):
  return 1/(1+np.exp(-w*x-b))
```

```python
x = np.linspace(-2,2,50)

y1 = sigmoid(1,0,x)
y2 = sigmoid(1,5,x)
y3 = sigmoid(1,10,x)
```

```python
fig, ax = plt.subplots()
ax.plot(x,y1,'-r', label='w=1, b =0')
ax.plot(x,y2,'-g', label='w=1, b =5')
ax.plot(x,y3,'-b', label='w=1, b =10')
leg=ax.legend()
```

```python
y1 = sigmoid(10,0,x)
y2 = sigmoid(10,5,x)
y3 = sigmoid(10,15,x)

fig, ax = plt.subplots()
ax.plot(x,y1,'-r', label='w=10, b =0')
ax.plot(x,y2,'-g', label='w=10, b =5')
ax.plot(x,y3,'-b', label='w=10, b =15')
leg=ax.legend()
```

```python
y1 = sigmoid(10,5,x)
y2 = sigmoid(20,10,x)
y3 = sigmoid(40,20,x)

fig, ax = plt.subplots()
ax.plot(x,y1,'-r', label='w=10, b =5')
ax.plot(x,y2,'-g', label='w=20, b =10')
ax.plot(x,y3,'-b', label='w=40, b =20')
leg=ax.legend()
```
