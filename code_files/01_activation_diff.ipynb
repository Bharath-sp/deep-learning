```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#activation functions

def sigmoid(x): return 1.0 / (1.0 + np.exp(-x))
def relu(x): return np.maximum(0,x)
def tanh(x): return np.tanh(x)
def softplus(x): return np.log(1 + np.exp(x))
```

```python
#numerical differentiation

def num_differentiation(f, x, h):
    return (f(x+h) - f(x-h))/ (2.0 * h)
```

```python
x = np.linspace(-5,5,100)
h = 1e-08
f = sigmoid
```

```python
fx = f(x)
```

```python
gx = num_differentiation(f, x, h)
```

$f(x) = \frac{1}{1 + e^{-x}}$. If we differentiate this function mathematically, we get $g(x) = f'(x)= f(x)(1-f(x))$. The graph of the numerical differentiation is similar to that of the mathematical differentiation.

```python
#visualization
plt.plot(x, fx, label='function f(x)')
plt.plot(x, gx, label='derivative g(x)')
plt.legend()
plt.show()
```

The ReLU function and its derivative is

$$
f(x) = \begin{cases}
        0 & \text{if } x \leq 0\\
        x & \text{if } x > 0
    \end{cases} \hspace{1cm} f'(x) = \begin{cases}
        0 & \text{if } x \leq 0\\
        1 & \text{if } x > 0
    \end{cases}
$$

It's derivative is undefined at $x=0$.

```python
f = relu
fx = f(x)
gx = num_differentiation(f, x, h)

plt.plot(x, fx, label='function f(x)')
plt.plot(x, gx, label='derivative g(x)')
plt.legend()
plt.show()
```

$f(x) = tanh(x), \hspace{1cm} g(x) = f'(x)= 1-tanh^2(x)$

```python
f = tanh
fx = f(x)
gx = num_differentiation(f, x, h)

plt.plot(x, fx, label='function f(x)')
plt.plot(x, gx, label='derivative g(x)')
plt.legend()
plt.show()
```

$f(x)= log(1+e^x), \hspace{1cm} g(x)=f'(x)=\frac{1}{1+e^{-x}}$

```python
f = softplus
fx = f(x)
gx = num_differentiation(f, x, h)

plt.plot(x, fx, label='function f(x)')
plt.plot(x, gx, label='derivative g(x)')
plt.legend()
plt.show()
```
