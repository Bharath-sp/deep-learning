```python
import numpy as np
```

```python
def relu(x):
  if x >0:
    return x
  else:
    return 0
```

```python
num_elements = 500
x = np.linspace(-10,10,num_elements , endpoint = False)
y = np.zeros(num_elements)
for i in range(num_elements):
  y[i] = relu(x[i])
```

```python
from matplotlib import pyplot as plt
```

```python
plt.plot(x,y,'-r')
plt.show()
```

```python
y_ = np.zeros(num_elements)
for i in range(num_elements):
  y_[i] = relu(-1*x[i])
```

```python
plt.plot(x,y_,'-r')
plt.show()
```

```python
plt.plot(x,y+y_,'-r')
plt.xlim(-3,3)
plt.show()
```

```python
# relu(x+1)
y1 = np.zeros(num_elements)
for i in range(num_elements):
  y1[i] = relu(1*x[i]+0)
# relu(-x+1)
y2 = np.zeros(num_elements)
for i in range(num_elements):
  y2[i] = relu(1*x[i]-0.1)

f = plt.figure(figsize=(15,4))

ax = f.add_subplot(132)
ax.set_xlim([-5,2])
ax.set_ylim([-5,5])

ax2= f.add_subplot(133)
ax2.set_xlim([-5,5])
ax2.set_ylim([-5,5])


ax.plot(x,y1,'-r')
ax.plot(x,y2,'-b')
ax2.plot(x,10*(y1-y2),'-r')
```

```python
def lin(x):
  return x
```

```python
# relu(x+1)
y3 = np.zeros(num_elements)
for i in range(num_elements):
  y3[i] = lin(1*x[i]+0)
# relu(-x+1)
y4 = np.zeros(num_elements)
for i in range(num_elements):
  y4[i] = lin(1*x[i]-1)

f = plt.figure(figsize=(15,4))

ax2 = f.add_subplot(132)
ax2.set_xlim([-5,5])
ax2.set_ylim([-5,5])

ax3= f.add_subplot(133)
ax3.set_xlim([-5,5])
ax3.set_ylim([-5,5])

ax2.plot(x,y3,'-r')
ax2.plot(x,y4,'-b')
ax3.plot(x,y3-y4,'-r')
```
