```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5,7,50)
```

```python
 # Quadratic function 2x^2-4x+2
def f(x):
  return 2*(x-1)**2
```

```python
# Derivatives of our quadratic function

def df(x_k):
  return 4*(x_k-1)

def df2(x_k):
  return 4
```

```python
### plot the quadratic function ###
y = f(x)
fig, ax1 = plt.subplots()

ax1.plot(x, y, 'g-')
ax1.set_ylabel('f(x)', color='g')
```

```python
### plot the quadratic function ###
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(x, y, 'g-')
ax1.set_ylabel('f(x)', color='g')

x0 = -3.5

ax1.scatter(x0,f(x0))
```

```python
### plot the quadratic function ###
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(x, y, 'g-')
ax1.set_ylabel('f(x)', color='g')

ax1.scatter(x0,f(x0))

eta = 0.1
x1 = x0 - eta*(df(x0))
ax1.scatter(x1,f(x1))
ax1.arrow(x0,f(x0), x1-x0, f(x1)-f(x0),width=0.085,shape='full',facecolor="red")
```
